```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Ellipse
import matplotlib.patches as patches
```

```python

def basic_pole_polar_concept():
    """Show the basic pole-polar relationship with a circle"""
    
    print("🎯 POLE-POLAR RELATIONSHIP: The Simple Version")
    print("=" * 55)
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    # Case 1: Point outside circle
    ax1 = axes[0]
    circle1 = Circle((0, 0), 2, fill=False, color='blue', linewidth=2)
    ax1.add_patch(circle1)
    
    # Pole outside circle
    pole_outside = np.array([3, 1])
    ax1.plot(pole_outside[0], pole_outside[1], 'ro', markersize=8, label='Pole (outside)')
    
    # Draw tangent lines from pole to circle
    # For circle x² + y² = r², tangent from point (a,b) has equation:
    # ax + by = r²
    r = 2
    a, b = pole_outside[0], pole_outside[1]
    
    # Find tangent points using geometry
    d = np.sqrt(a*a + b*b)  # distance from origin to pole
    l = np.sqrt(d*d - r*r)  # length of tangent
    
    # Angle calculations for tangent points
    theta_pole = np.arctan2(b, a)
    alpha = np.arcsin(r/d)
    
    # Two tangent points
    theta1 = theta_pole + alpha
    theta2 = theta_pole - alpha
    
    t1 = np.array([r * np.cos(theta1), r * np.sin(theta1)])
    t2 = np.array([r * np.cos(theta2), r * np.sin(theta2)])
    
    # Draw tangent lines
    ax1.plot([pole_outside[0], t1[0]], [pole_outside[1], t1[1]], 'g--', alpha=0.7)
    ax1.plot([pole_outside[0], t2[0]], [pole_outside[1], t2[1]], 'g--', alpha=0.7)
    ax1.plot([t1[0], t2[0]], [t1[1], t2[1]], 'r-', linewidth=3, label='Polar line')
    
    ax1.set_xlim(-3, 4)
    ax1.set_ylim(-3, 3)
    ax1.grid(True)
    ax1.set_aspect('equal')
    ax1.legend()
    ax1.set_title('Case 1: Pole Outside Circle\nPolar = Line connecting tangent points')
    
    # Case 2: Point on circle
    ax2 = axes[1]
    circle2 = Circle((0, 0), 2, fill=False, color='blue', linewidth=2)
    ax2.add_patch(circle2)
    
    # Pole on circle
    pole_on = np.array([2*np.cos(np.pi/4), 2*np.sin(np.pi/4)])  # Point on circle
    ax2.plot(pole_on[0], pole_on[1], 'ro', markersize=8, label='Pole (on circle)')
    
    # Polar is tangent at that point
    # Tangent line at point (x₀, y₀) on circle x² + y² = r² is: x₀x + y₀y = r²
    x_range = np.linspace(-3, 3, 100)
    # y = (r² - x₀*x) / y₀
    y_tangent = (4 - pole_on[0] * x_range) / pole_on[1]
    
    # Only plot reasonable range
    valid_idx = (y_tangent >= -3) & (y_tangent <= 3)
    ax2.plot(x_range[valid_idx], y_tangent[valid_idx], 'r-', linewidth=3, label='Polar line (tangent)')
    
    ax2.set_xlim(-3, 3)
    ax2.set_ylim(-3, 3)
    ax2.grid(True)
    ax2.set_aspect('equal')
    ax2.legend()
    ax2.set_title('Case 2: Pole On Circle\nPolar = Tangent line at that point')
    
    # Case 3: Point inside circle
    ax3 = axes[2]
    circle3 = Circle((0, 0), 2, fill=False, color='blue', linewidth=2)
    ax3.add_patch(circle3)
    
    # Pole inside circle
    pole_inside = np.array([0.5, 0.8])
    ax3.plot(pole_inside[0], pole_inside[1], 'ro', markersize=8, label='Pole (inside)')
    
    # For point inside, polar is the line that would be tangent line if the point were reflected outside
    # Mathematical formula: for pole (a,b) and circle x² + y² = r², polar line is: ax + by = r²
    a, b = pole_inside[0], pole_inside[1]
    r = 2
    
    # Polar line: ax + by = r²  →  y = (r² - ax)/b
    x_range = np.linspace(-3, 3, 100)
    y_polar = (r*r - a * x_range) / b
    
    valid_idx = (y_polar >= -3) & (y_polar <= 3)
    ax3.plot(x_range[valid_idx], y_polar[valid_idx], 'r-', linewidth=3, label='Polar line')
    
    ax3.set_xlim(-3, 3)
    ax3.set_ylim(-3, 3)
    ax3.grid(True)
    ax3.set_aspect('equal')
    ax3.legend()
    ax3.set_title('Case 3: Pole Inside Circle\nPolar = Line with special relationship')
    
    plt.tight_layout()
    plt.show()
    
    print("🔍 SIMPLE UNDERSTANDING:")
    print("• POLE = A point")
    print("• POLAR = A line that has a special relationship with that point")
    print("• The relationship depends on where the pole is relative to the conic")
    print()
    print("📐 MATHEMATICAL FORMULA:")
    print("For a conic C and pole point p, the polar line l satisfies: l = C⁻¹p")
    print("(Don't worry about the math yet - focus on the visual relationship!)")
```

```python

def demonstrate_reciprocal_property():
    """Show the amazing reciprocal property"""
    
    print("\n✨ THE AMAZING RECIPROCAL PROPERTY")
    print("=" * 50)
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 8))
    
    # Draw circle
    circle = Circle((0, 0), 2, fill=False, color='blue', linewidth=2)
    ax.add_patch(circle)
    
    # Choose two points
    point_A = np.array([3, 1])
    point_B = np.array([1, 2.5])
    
    ax.plot(point_A[0], point_A[1], 'ro', markersize=10, label='Point A')
    ax.plot(point_B[0], point_B[1], 'go', markersize=10, label='Point B')
    
    # Calculate polar of A
    # For circle x² + y² = 4, polar of (a,b) is ax + by = 4
    r_sq = 4
    a_A, b_A = point_A[0], point_A[1]
    a_B, b_B = point_B[0], point_B[1]
    
    # Polar of A: a_A * x + b_A * y = 4
    x_range = np.linspace(-1, 4, 100)
    y_polar_A = (r_sq - a_A * x_range) / b_A
    
    # Polar of B: a_B * x + b_B * y = 4  
    y_polar_B = (r_sq - a_B * x_range) / b_B
    
    # Plot polar lines
    valid_A = (y_polar_A >= -1) & (y_polar_A <= 4)
    valid_B = (y_polar_B >= -1) & (y_polar_B <= 4)
    
    ax.plot(x_range[valid_A], y_polar_A[valid_A], 'r--', linewidth=2, label='Polar of A')
    ax.plot(x_range[valid_B], y_polar_B[valid_B], 'g--', linewidth=2, label='Polar of B')
    
    # Check if B lies on polar of A
    # Substitute B into equation of polar of A
    result_B_on_polar_A = a_A * point_B[0] + b_A * point_B[1]
    print(f"Point B coordinates: ({point_B[0]}, {point_B[1]})")
    print(f"Polar of A equation: {a_A}x + {b_A}y = {r_sq}")
    print(f"Substituting B into polar of A: {a_A}×{point_B[0]} + {b_A}×{point_B[1]} = {result_B_on_polar_A:.2f}")
    print(f"Expected value: {r_sq}")
    
    # Check if A lies on polar of B
    result_A_on_polar_B = a_B * point_A[0] + b_B * point_A[1] 
    print(f"Substituting A into polar of B: {a_B}×{point_A[0]} + {b_B}×{point_A[1]} = {result_A_on_polar_B:.2f}")
    
    # Mark intersection points
    if abs(result_B_on_polar_A - r_sq) < 0.1:
        ax.plot(point_B[0], point_B[1], 'ro', markersize=15, fillstyle='none', linewidth=3)
        ax.annotate('B lies on polar of A!', xy=point_B, xytext=(point_B[0]+0.3, point_B[1]+0.3),
                   arrowprops=dict(arrowstyle='->', color='red'))
    
    if abs(result_A_on_polar_B - r_sq) < 0.1:
        ax.plot(point_A[0], point_A[1], 'go', markersize=15, fillstyle='none', linewidth=3)
        ax.annotate('A lies on polar of B!', xy=point_A, xytext=(point_A[0]+0.3, point_A[1]-0.5),
                   arrowprops=dict(arrowstyle='->', color='green'))
    
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 4)
    ax.grid(True)
    ax.set_aspect('equal')
    ax.legend()
    ax.set_title('Reciprocal Property: If B lies on polar of A,\nthen A lies on polar of B!')
    
    plt.tight_layout()
    plt.show()
    
    print("\n🎯 RECIPROCAL PROPERTY:")
    print("If point B lies on the polar line of point A,")
    print("then point A MUST lie on the polar line of point B!")
    print("This is a fundamental symmetry in projective geometry.")

# ============================================================================
# PART 2: REAL-WORLD APPLICATIONS
# ============================================================================

def application_vanishing_points():
    """Show how pole-polar relates to vanishing points"""
    
    print("\n🌍 REAL-WORLD APPLICATION 1: VANISHING POINTS")
    print("=" * 55)
    
    print("📸 CONNECTION TO COMPUTER VISION:")
    print("The pole-polar relationship is fundamental to understanding vanishing points!")
    print()
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # Left: 3D scene with parallel lines
    ax1.set_title('3D Scene: Parallel Lines')
    
    # Draw parallel lines in 3D (viewed from above)
    line1_start = np.array([0, 1])
    line1_end = np.array([5, 1])
    line2_start = np.array([0, -1]) 
    line2_end = np.array([5, -1])
    
    ax1.plot([line1_start[0], line1_end[0]], [line1_start[1], line1_end[1]], 'b-', linewidth=3, label='Parallel line 1')
    ax1.plot([line2_start[0], line2_end[0]], [line2_start[1], line2_end[1]], 'r-', linewidth=3, label='Parallel line 2')
    
    # Show direction
    ax1.arrow(2.5, 0, 1, 0, head_width=0.2, head_length=0.2, fc='green', ec='green')
    ax1.text(3, 0.3, 'Direction of\nparallel lines', ha='center', fontsize=10, 
            bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen"))
    
    ax1.set_xlim(-0.5, 6)
    ax1.set_ylim(-2, 2)
    ax1.grid(True)
    ax1.legend()
    ax1.set_aspect('equal')
    
    # Right: 2D image with vanishing point
    ax2.set_title('2D Image: Lines Meet at Vanishing Point')
    
    # Simulate perspective projection - lines converge
    img_line1_start = np.array([1, 2])
    img_line1_end = np.array([4, 1.2])
    img_line2_start = np.array([1, -2])
    img_line2_end = np.array([4, -1.2])
    
    # Extend lines to find vanishing point
    x_extend = np.linspace(1, 8, 100)
    # Line 1: slope = (1.2-2)/(4-1) = -0.8/3
    slope1 = (1.2 - 2) / (4 - 1)
    y1_extend = 2 + slope1 * (x_extend - 1)
    
    slope2 = (-1.2 - (-2)) / (4 - 1)
    y2_extend = -2 + slope2 * (x_extend - 1)
    
    ax2.plot(x_extend, y1_extend, 'b-', linewidth=2, label='Line 1 in image')
    ax2.plot(x_extend, y2_extend, 'r-', linewidth=2, label='Line 2 in image')
    
    # Mark vanishing point
    # Solve for intersection: 2 + slope1*(x-1) = -2 + slope2*(x-1)
    vp_x = 1 + (4) / (slope1 - slope2)
    vp_y = 2 + slope1 * (vp_x - 1)
    
    ax2.plot(vp_x, vp_y, 'ko', markersize=10, label='Vanishing Point')
    ax2.annotate('Vanishing Point\n(pole of line at infinity)', 
                xy=(vp_x, vp_y), xytext=(vp_x-1, vp_y+1),
                arrowprops=dict(arrowstyle='->', color='black'),
                fontsize=10, ha='center',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow"))
    
    ax2.set_xlim(0, 8)
    ax2.set_ylim(-3, 3)
    ax2.grid(True)
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    print("🔗 THE CONNECTION:")
    print("• In 3D: Parallel lines have a common direction")
    print("• This direction is a 'point at infinity' in projective geometry")
    print("• In 2D image: This 'point at infinity' becomes the vanishing point")
    print("• The vanishing point is the POLE of the 'line at infinity'")
    print("• Pole-polar relationship explains why parallel lines meet at vanishing points!")
```

```python
def application_camera_calibration():
    """Show application in camera calibration"""
    
    print("\n🌍 REAL-WORLD APPLICATION 2: CAMERA CALIBRATION")
    print("=" * 55)
    
    print("📐 ABSOLUTE CONIC & CAMERA CALIBRATION:")
    print("The pole-polar relationship with the 'absolute conic' is used to calibrate cameras!")
    print()
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
    
    # Left: Show the concept
    ax1.set_title('Absolute Conic Concept')
    
    # Draw the "absolute conic" (conceptual - it's actually at infinity)
    circle = Circle((0, 0), 2, fill=False, color='purple', linewidth=3, linestyle='--')
    ax1.add_patch(circle)
    ax1.text(0, 0, 'Absolute Conic\n(at infinity)', ha='center', va='center', 
            fontsize=12, weight='bold',
            bbox=dict(boxstyle="round,pad=0.5", facecolor="purple", alpha=0.7))
    
    # Show orthogonal directions
    ax1.arrow(0, 0, 1.5, 0, head_width=0.1, head_length=0.1, fc='red', ec='red')
    ax1.arrow(0, 0, 0, 1.5, head_width=0.1, head_length=0.1, fc='blue', ec='blue')
    ax1.text(1.7, 0, 'Direction 1', fontsize=10, color='red')
    ax1.text(0.1, 1.7, 'Direction 2', fontsize=10, color='blue', rotation=90)
    
    # Show that orthogonal directions are pole-polar with respect to absolute conic
    ax1.plot([0, 1.5], [0, 0], 'r-', linewidth=2)
    ax1.plot([0, 0], [0, 1.5], 'b-', linewidth=2)
    
    ax1.set_xlim(-3, 3)
    ax1.set_ylim(-3, 3)
    ax1.grid(True)
    ax1.set_aspect('equal')
    
    # Right: Practical application
    ax2.set_title('Camera Calibration Application')
    
    # Simulate a checkerboard pattern image
    # Draw perspective view of checkerboard
    corners = np.array([
        [1, 1], [3, 1.2], [5, 1.5],  # Top row
        [1, 2], [3, 2.1], [5, 2.3],  # Middle row  
        [1, 3], [3, 3.0], [5, 3.2]   # Bottom row
    ])
    
    # Draw checkerboard grid
    for i in range(3):
        for j in range(3):
            idx = i * 3 + j
            ax2.plot(corners[idx, 0], corners[idx, 1], 'ko', markersize=6)
            
    # Connect to show grid
    for i in range(3):
        # Horizontal lines
        row_start = i * 3
        ax2.plot([corners[row_start, 0], corners[row_start + 2, 0]], 
                [corners[row_start, 1], corners[row_start + 2, 1]], 'k--', alpha=0.5)
        
        # Vertical lines  
        ax2.plot([corners[i, 0], corners[i + 6, 0]], 
                [corners[i, 1], corners[i + 6, 1]], 'k--', alpha=0.5)
    
    # Mark vanishing points
    vp1 = np.array([7, 2])  # Horizontal vanishing point
    vp2 = np.array([3, 0])  # Vertical vanishing point
    
    ax2.plot(vp1[0], vp1[1], 'r^', markersize=10, label='Vanishing Point 1')
    ax2.plot(vp2[0], vp2[1], 'b^', markersize=10, label='Vanishing Point 2')
    
    # Draw lines to vanishing points
    ax2.plot([1, vp1[0]], [1, vp1[1]], 'r:', alpha=0.7)
    ax2.plot([1, vp2[0]], [1, vp2[1]], 'b:', alpha=0.7)
    
    ax2.set_xlim(0, 8)
    ax2.set_ylim(0, 4)
    ax2.grid(True)
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    print("🔧 HOW IT WORKS:")
    print("1. Take photos of rectangular patterns (checkerboards)")
    print("2. Find vanishing points of orthogonal line sets")
    print("3. Use pole-polar relationship with absolute conic")
    print("4. Extract camera intrinsic parameters (focal length, etc.)")
    print()
    print("💡 KEY INSIGHT:")
    print("Orthogonal directions in 3D world are pole-polar pairs")
    print("with respect to the absolute conic. This constraint")
    print("allows us to calibrate the camera without knowing")
    print("the 3D coordinates of the calibration pattern!")
```

```python

def application_epipolar_geometry():
    """Show application in stereo vision"""
    
    print("\n🌍 REAL-WORLD APPLICATION 3: STEREO VISION")
    print("=" * 55)
    
    print("👁️👁️ EPIPOLAR GEOMETRY:")
    print("Pole-polar relationship explains epipolar lines in stereo vision!")
    print()
    
    fig, ax = plt.subplots(1, 1, figsize=(12, 8))
    
    # Draw two cameras
    cam1_pos = np.array([1, 3])
    cam2_pos = np.array([5, 3])
    
    ax.plot(cam1_pos[0], cam1_pos[1], 'bs', markersize=12, label='Camera 1')
    ax.plot(cam2_pos[0], cam2_pos[1], 'rs', markersize=12, label='Camera 2')
    
    # Draw image planes
    img1_x = [cam1_pos[0], cam1_pos[0]]
    img1_y = [cam1_pos[1] - 1, cam1_pos[1] + 1]
    img2_x = [cam2_pos[0], cam2_pos[0]]
    img2_y = [cam2_pos[1] - 1, cam2_pos[1] + 1]
    
    ax.plot(img1_x, img1_y, 'b-', linewidth=3, label='Image 1')
    ax.plot(img2_x, img2_y, 'r-', linewidth=3, label='Image 2')
    
    # 3D point
    point_3d = np.array([3, 1])
    ax.plot(point_3d[0], point_3d[1], 'go', markersize=10, label='3D Point')
    
    # Project point to both cameras
    # Simplified projection (just lines from cameras to point)
    ax.plot([cam1_pos[0], point_3d[0]], [cam1_pos[1], point_3d[1]], 'g--', alpha=0.7)
    ax.plot([cam2_pos[0], point_3d[0]], [cam2_pos[1], point_3d[1]], 'g--', alpha=0.7)
    
    # Find projections on image planes (simplified)
    # For camera 1: intersection with vertical line at x=1
    proj1_y = cam1_pos[1] + (point_3d[1] - cam1_pos[1]) * (cam1_pos[0] - cam1_pos[0]) / (point_3d[0] - cam1_pos[0])
    proj1 = np.array([cam1_pos[0], proj1_y])
    
    # For camera 2: intersection with vertical line at x=5  
    proj2_y = cam2_pos[1] + (point_3d[1] - cam2_pos[1]) * (cam2_pos[0] - cam2_pos[0]) / (point_3d[0] - cam2_pos[0])
    proj2 = np.array([cam2_pos[0], proj2_y])
    
    ax.plot(proj1[0], proj1[1], 'bo', markersize=8, label='Projection in Image 1')
    ax.plot(proj2[0], proj2[1], 'ro', markersize=8, label='Projection in Image 2')
    
    # Show epipolar line concept
    # Draw baseline (line connecting camera centers)
    ax.plot([cam1_pos[0], cam2_pos[0]], [cam1_pos[1], cam2_pos[1]], 'k-', linewidth=2, label='Baseline')
    
    # Epipoles (where baseline intersects image planes)
    epipole1 = np.array([cam1_pos[0], cam1_pos[1]])
    epipole2 = np.array([cam2_pos[0], cam2_pos[1]])
    
    ax.plot(epipole1[0], epipole1[1], 'ko', markersize=8, label='Epipole 1')
    ax.plot(epipole2[0], epipole2[1], 'ko', markersize=8, label='Epipole 2')
    
    # Epipolar lines (lines from epipoles through projections)
    ax.plot([epipole1[0], proj1[0]], [epipole1[1], proj1[1]], 'm--', linewidth=2, label='Epipolar line 1')
    ax.plot([epipole2[0], proj2[0]], [epipole2[1], proj2[1]], 'm--', linewidth=2, label='Epipolar line 2')
    
    ax.set_xlim(0, 6)
    ax.set_ylim(0, 5)
    ax.grid(True)
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.set_title('Epipolar Geometry: Pole-Polar Relationship in Stereo Vision')
    
    plt.tight_layout()
    plt.show()
    
    print("🔗 THE POLE-POLAR CONNECTION:")
    print("• Each point in one image is the POLE")
    print("• The corresponding epipolar line in the other image is its POLAR")
    print("• This relationship is defined by the fundamental matrix")
    print("• Used for stereo matching and 3D reconstruction")
    print()
    print("🎯 PRACTICAL USE:")
    print("• Stereo vision algorithms use this to constrain the search")
    print("• Instead of searching the entire second image for a match,")
    print("• only search along the epipolar line (1D instead of 2D search!)")
```
